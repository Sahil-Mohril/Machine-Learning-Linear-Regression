```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols=np.loadtxt("ex1data1.txt",delimiter=",",usecols=(0,1),unpack=True)
xt=np.array(cols[0])
yt=np.array(cols[1])
X=np.transpose(np.array(cols[:-1]))
y=np.transpose(np.array(cols[-1:]))
m=y.size
X=np.insert(X,0,1,axis=1)
```

```python
from scipy.stats import linregress
fit=linregress(xt,yt)
```

```python
y_pred=fit.intercept+fit.slope*xt
plt.scatter(xt, yt, color='blue', label='Data points')
slope=fit.slope
intercept=fit.intercept
# Plotting the best fit line
plt.plot(xt, y_pred, color='red', label=f'Best fit line: y={slope:.2f}x+{intercept:.2f}')
plt.legend()
```
