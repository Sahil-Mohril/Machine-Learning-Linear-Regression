```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cols=np.loadtxt("ex1data1.txt",delimiter=",",usecols=(0,1),unpack=True)
X=np.transpose(np.array(cols[:-1]))
y=np.transpose(np.array(cols[-1:]))
m=y.size
m
X=np.insert(X,0,1,axis=1)
```

```python

plt.plot(X[:,1],y[:,0],'rx',markersize=5)
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
```

```python
#Cost Function
def cost_function(theta,X,y):
    return (1.0/(2*m))*np.sum((np.dot(X,theta)-y)**2)
cost_function(np.zeros((X.shape[1],1)),X,y)
```

```python
theta=np.zeros((X.shape[1],1))
jvec=[]
for epoch in range(1500):
    for j in range(len(theta)):
        jvec.append(cost_function(theta,X,y))
        theta[j]=theta[j]-(0.01/m)*np.sum((np.dot(X,theta)-y)*np.array(X[:,j]).reshape(m,1))
theta
```

```python
def myfit(xval):
    return theta[0] + theta[1]*xval
plt.plot(X[:,1],y[:,:],'rx',markersize=5,label='Training Data')
plt.plot(X[:,1],myfit(X[:,1]),'b-',label = 'Hypothesis: h(x) = %0.2f + %0.2fx'%(theta[0],theta[1]))

plt.ylabel('Profit in $10,000s')
plt.legend()
```

```python
plt.figure(figsize=(5,4))
plt.plot(range(len(jvec)),jvec,'k')
plt.xlim([-0.05*1500,1.08*1500])
plt.ylim(4,10)
plt.title("Convergence of Cost Function")
plt.ylabel('Cost value')
plt.xlabel('iteration')
```
